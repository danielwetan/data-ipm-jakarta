Import library yang diperlukan

```python
import folium
import pandas as pd
```

Generate simple map

```python
m = folium.Map(location=[-6.2049, 106.8633], zoom_start=11)
m
```

Highlight map pada wilayah DKI Jakarta

```python
# import data polygon dari kabupaten di DKI Jakarta
data_kabupaten = 'data-kabupaten-jakarta.json'

# inisialisasi map folium
m = folium.Map(location=[-6.2049, 106.8633], zoom_start=11)

# highlight map berdasarkan data polygon kabupaten
folium.GeoJson(data_kabupaten).add_to(m)

# render map
m
```

Import data Indeks Pembangunan Manusia (IPM) DKI Jakarta

```python
data_ipm = pd.read_csv('data-ipm-jakarta.csv')
data_ipm.head()

# note: data yang ada pada kolom city merupakan id dari kabupaten yang ada di Jakarta
# ID3171: Jakarta Selatan
# ID3172: Jakarta Timur
# ID3173: Jakarta Pusat
# ID3174: Jakarta Barat
# ID3175: Jakarta Utara
```

Tampilkan map Indeks Pembangunan Manusia (IPM) DKI Jakarta

```python
# inisialisasi map folium
m = folium.Map(location=[-6.2049, 106.8633], zoom_start=11)

# ubah TileLayer pada map menjadi cartodbpositron
folium.TileLayer('cartodbpositron').add_to(m)

# render map berdasarkan data ipm dan kabupaten
folium.Choropleth(
    geo_data=data_kabupaten,
    name='choropleth',
    data=data_ipm,
    columns=['city', 'ipm'],
    key_on='feature.id',
    fill_color='YlGn',
    fill_opacity=0.7,
    line_opacity=0.2,
    legend_name='IPM Rate'
).add_to(m)
m
```

Simpan map ke file HTML

```python
m.save('output_map.html')
```
